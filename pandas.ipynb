```python
import numpy as np
import pandas as pd
```

```python
#statistical operations
s=pd.Series([1,2,6,3,5])
print(s)
print(s.describe())
print(s.mean())
print(s.median())
print(s.std())
```

```python
#data manipulation on series
print(s.map(lambda x:x+2))
print(s.apply(lambda x:x**2))
print(s.sort_values())
print(s.drop(3))
```

```python
#handling data on series:

s1=pd.Series([4,6,3,8,None,1,np.nan])
print(s1)
print(s1.isnull())
print(s1.notnull())
print(s1.fillna(0))
print(s1.dropna())
```

```python
#indexing,slicing and filtering
s2=pd.Series([4,3,6,2,1,5])
print(s2.iloc[2])
print(s2.iloc[-2])
# print(s2.loc[3])
#print(s2[s2>=1])
```

```python
#aggregation:
print(s2.sum())
print(s2.cumsum())
print(s.aggregate(['sum','std','mean']))
```

```python
#DATAFRAMES:

#Creating dataframes:

data1={
    "name":['chikki','chinni','teju','meghna','anu','spandy','aishu'],
    "age":[1,22,20,23,24,16,14],
    "branch":['cse','ece','it','eee','aiml','ds','cs'],
    "address":['hyd','amgl','shkpt','dsnr','jrp','krmgt','chmpt']
    }
df=pd.DataFrame(data1)
print(df)
print(df.head(2))
print(df.tail(2))
```

```python
#combining data from two dataframes

data1={
    "name":['chikki','chinni','teju','meghna','anu','spandy','aishu'],
    "age":[21,22,20,23,24,16,14],
    "branch":['cse','ece','it','eee','aiml','ds','cs']
    }
data2={
    "name":['chikki','chinni','teju','meghna','anu','spandy','aishu'],
    "age":[21,22,20,23,24,16,14],
    "address":['hyd','amgl','shkpt','dsnr','jrp','krmgt','chmpt']
    }
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
merged=pd.merge(df1,df2,on="name")
print(merged)
```
