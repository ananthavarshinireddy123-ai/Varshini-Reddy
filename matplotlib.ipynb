```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#line plot
x=[1,2,3,4,5,6]
y=[10,20,30,40,50,60]
plt.plot(x,y)
plt.title('lineplot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
#bar garph
student=['A','B','C','D','E']
marks=[97.96,80,75,35,20]
plt.bar(student,marks)
plt.title('bargraph')
plt.xlabel('student')
plt.ylabel('marks')
plt.show()
```

```python
ages=[20,30,40,50,60,70,99,101]
plt.hist(ages,bins=5)
plt.title('age distribution')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
time=[1,2,3,4,5,6]
result=[10,20,40,60,80,100]
plt.scatter(time,result)
plt.title('anaylsis')
plt.xlabel('time')
plt.ylabel('result')
plt.legend(['result'])
plt.show()
```

```python
subjects=['english','maths','science','social']
marks=[40,10,30,20]
plt.pie(marks,labels=subjects,autopct='%1.1f%%')
plt.title('sub analysis')
plt.show()
```
