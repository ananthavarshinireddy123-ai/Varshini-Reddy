#Introduction to Reinforcement Learning (RL)

Reinforcement Learning (RL) is a branch of machine learning in which an agent learns how to make decisions by interacting with an environment. The agent learns through trial and error, receiving rewards for good actions and penalties for bad actions. Over time, it discovers the best sequence of actions to maximize its total reward.

Unlike supervised learning, where the model learns from labeled data, reinforcement learning does not require labeled examples. Instead, it learns from feedback in the form of rewards.

#Simple Example

Imagine teaching a dog to sit:

If the dog sits when commanded, you give it a treat (reward).

If it doesn't sit, it gets no treat (penalty or no reward).

After repeating this process many times, the dog learns that sitting leads to a reward.

Similarly, an RL agent learns which actions produce the highest rewards.

| Component       | Description                           | Example                                          |
| --------------- | ------------------------------------- | ------------------------------------------------ |
| **Agent**       | The learner or decision-maker         | Robot                                            |
| **Environment** | The world in which the agent operates | Maze                                             |
| **State (S)**   | Current situation of the agent        | Robot's current position                         |
| **Action (A)**  | Choice made by the agent              | Move left, right, up, or down                    |
| **Reward (R)**  | Feedback received after an action     | +10 for reaching the goal, -1 for hitting a wall |
| **Policy (π)**  | Strategy used to choose actions       | Rules for selecting the next move                |

#How Reinforcement Learning Works
1.The agent observes the current state.

2.It selects an action.

3.The environment changes to a new state.

4.The environment provides a reward.

5.The agent updates its strategy based on the reward.

6.The process repeats until the agent learns the best policy.

#Applications of Reinforcement Learning
-Game playing (Chess, Go, Atari games)

-Self-driving cars

-Robotics and automation

-Stock trading

-Recommendation systems

-Network traffic optimization

#Advantages
Learns without labeled data.

Improves through continuous interaction.

Solves complex sequential decision-making problems.
Can adapt to changing environments.

#Limitations
Requires many interactions for training.

Training can be computationally expensive.

Designing an effective reward function can be difficult.

May explore inefficient actions before learning the optimal policy.

Example 1: Simple Reward-Based Learning

This example demonstrates the basic idea of reinforcement learning.

```python
import random

actions = ["Left", "Right"]

for episode in range(10):
    action = random.choice(actions)

    if action == "Right":
        reward = 10
    else:
        reward = -5

    print(f"Episode {episode+1}")
    print("Action:", action)
    print("Reward:", reward)
    print()
```

#Markov Decision Process (MDP) in Machine Learning

A Markov Decision Process (MDP) is a mathematical framework used in Reinforcement Learning (RL) to model decision-making problems where an agent interacts with an environment to maximize rewards.

The Markov Property states that the future depends only on the current state, not on the sequence of previous states.

For example, in a maze game, if a robot is currently at position A, the next move depends only on its current position (A) and not on how it reached A.

| Component                      | Meaning                                 | Example                                     |
| ------------------------------ | --------------------------------------- | ------------------------------------------- |
| **S (States)**                 | All possible situations                 | Room A, Room B, Goal                        |
| **A (Actions)**                | Choices available                       | Left, Right, Up, Down                       |
| **P (Transition Probability)** | Probability of moving to the next state | 80% chance of moving right                  |
| **R (Reward)**                 | Feedback after taking an action         | +10 for reaching the goal, -1 for each step |
| **γ (Gamma)**                  | Discount factor (0–1)                   | 0.9 gives importance to future rewards      |




```python
# States
states = ["Start", "Goal"]

# Rewards for each state
rewards = {
    "Start": -1,
    "Goal": 10
}

# Possible actions
actions = {
    "Start": "Move to Goal",
    "Goal": "Stop"
}

# Initial state
current_state = "Start"

print("Current State:", current_state)
print("Action:", actions[current_state])

# Transition to next state
current_state = "Goal"

print("\nNext State:", current_state)
print("Reward:", rewards[current_state])
```
