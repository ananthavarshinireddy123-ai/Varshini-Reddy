```python
import pandas as pd
import numpy as np
```

```python
data={
    "Name": ["chinni","chikki","varshini","teju",None,"anu"],
    "Age": [22,28, np.nan,21,30,25],
    "City": ["Hyd","Delhi","hyd","delhi","Delhi","Hyd"]
}

df=pd.DataFrame(data)
df
```

```python
# First few rows
print(df.head())

# Data info
print(df.info())

# Summary statistics
print(df.describe())
```

```python
# Check missing values
print(df.isnull())

# Count missing values
df.isnull().sum() #count of missing values
```

```python
# Drop rows with missing values
df_drop = df.dropna()
df_drop
```

```python
#  FIling Missing Values

df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Name"] = df["Name"].fillna(df["Name"].mode()[0])

df
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Age": [20, np.nan, 25, 30,20],
    "Salary": [20000, 25000, np.nan, 40000,20000],
    "City": ["Mumbai", "Delhi", "Pune", "Mumbai","Mumbai"]
})

df
```

```python
# Fill multiple columns with mean

df[["Age", "Salary"]] = df[["Age", "Salary"]].fillna(df[["Age", "Salary"]].mean())

df
```

```python
# Convert multiple columns to integer

df[["Age", "Salary"]] = df[["Age", "Salary"]].astype(int)

df.dtypes
df
```

```python
# Increase values by 10%

df[["Age", "Salary"]] = df[["Age", "Salary"]].apply(lambda x: x * 1.1)

df
```

```python
# Replace specific values

df[["City"]] = df[["City"]].replace("Mumbai", "MUMBAI")
df
```

```python
#Encoding
df_encoded=pd.get_dummies(df,columns=["City"])
df_encoded
```

```python
df_no_dup = df.drop_duplicates()
df_no_dup
```

```python
import pandas as pd

df = pd.read_csv("/content/rainfall.csv", encoding='latin1')

# First 5 rows
print(df.head())
# Last 5 rows
print(df.tail())
```

```python
df["YEAR"] = df["YEAR"].astype(str).str.strip()
df["YEAR"] = pd.to_datetime(df["YEAR"], errors="coerce")
df.info()
```

```python
print(df.head())
print(df.columns)
```

```python
import pandas as pd

# Numerical columns
num_cols = ["YEAR", "JAN", "FEB", "MAR", "APR", "MAY",
            "JUN", "JUL", "AUG", "SEP", "OCT", "NOV",
            "DEC", "ANNUAL", "Jan-Feb", "Mar-May",
            "Jun-Sep", "Oct-Dec"]

# Fill numerical missing values with mean
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# Fill categorical missing values with mode
df["SUBDIVISION"] = df["SUBDIVISION"].fillna(df["SUBDIVISION"].mode()[0])

# Check missing values
print(df.isnull().sum())
```

```python
print(df.head())
print(df.columns)
```

```python
import pandas as pd
import numpy as np

data = {
    "name": ["Amit", "Sara", "John", "Amit", None],
    "age": [20, 21, np.nan, 20, 19],
    "gender": ["Male", "female", "Male", "Male", "FEMALE"],
    "marks": [85, 90, 105, 85, -5],
    "attendance": [90, 85, 80, 90, None],
    "city": ["Mumbai", "Delhi", "mumbai", "Mumbai", "Pune"]
}

df = pd.DataFrame(data)
df
```

```python
# Fill missing values in numerical columns
df["age"] = df["age"].fillna(df["age"].mean())
df["attendance"] = df["attendance"].fillna(df["attendance"].mean())

# Fill missing values in categorical column
df["name"] = df["name"].fillna("Unknown")

# Display dataset
df
```

```python
df["gender"] = df["gender"].str.lower()
df["city"] = df["city"].str.lower()

df
```

```python
df = df[(df["marks"] >= 0) & (df["marks"] <= 100)]

df
```
