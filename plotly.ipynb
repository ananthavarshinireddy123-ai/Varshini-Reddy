```python
import pandas as pd
import plotly.express as px
data={
    'Department':['IT','HR','Finance','Marketing'],
    'Employees':[50,20,30,25]
    }
df=pd.DataFrame(data)
fig=px.pie(
    df,
    values='Employees',
    names='Department',
    title='Employee Distribution'
    )
fig.show()
```

```python
data={
    'Experience':[1,2,3,4,5,3,4,6],
    'Salary':[50000,60000,70000,80000,90000,65000,75000,95000]
    }
df=pd.DataFrame(data)
fig=px.scatter(
        df,
        x='Experience',
        y='Salary',
        title='Experience vs Salary'
        )
fig.show()
```

```python
fig=px.histogram(
    df,
    x='Experience',
    y='Salary',
    title='Experience vs Salary'
    )
fig.show()
```

```python
fig=px.bar(
    df,
    x='Experience',
    y='Salary',
    title='Experience vs Salary'
    )
fig.show()
```
